```python
# 판다 시작하기
# 리터널 : 정수 , 실수 , 문자열
# 부울
# 파이썬 : 리스트 , 튜플, 딕셔너리, 집합
# numpy : 숫자 배열 : 리스트
# pandas : 시리즈 : 1차원배열 : 벡터 : 리스트 : 값이 1개인 딕셔너리
#          DataFrame : 2차원배원 : 값이 2개이상인 딕셔너리
```

```python
import pandas as pd
# 시리즈
obj = pd.Series([4, 7, -5, 3])
print(obj)
print(obj[0])
print(obj.index)
```

```python
import numpy as np
arr = obj.to_numpy() # 1차원 배열
print(arr)
print(obj)
print(np.sum(obj))
print(np.sum(arr))
```

```python
obj = pd.Series([4, 7, -5, 3])
print(obj)
print(obj[0])
```

```python
obj = pd.Series([4, 7, -5, 3], index = ['a','b','c','d'])
print(obj)
print(obj['a']) # 명시적 index
# print(obj[0]) # 암시적 index
l = [1,2,3,4]
print(l[0])
#print(l[0,1,2])
arr =  np.array([1,2,3,4])
print(arr[0])
print(arr[[0,1,2]])
print(obj[['a','b','c']])
```

```python
print(arr)
print(arr[arr > 2]) # 부울 index
print(obj[obj > 2])
```

```python
# in 연산자
print( 'a' in obj) # obj에 index가 있는지 확인
print( 0 in obj)
print('e' in obj)
```

```python
# 딕셔너리를 이용한 시리즈
dic = obj.to_dict()
print(dic)
sdata = pd.Series({"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000})
print(sdata) # 딕셔너리의 키는 시리즈의 index가 됨
dic1 = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}
sdata1 = pd.Series(dic1)
print(sdata)
```

```python
dic1 = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}
sdata1 = pd.Series(dic1, index=["California", "Ohio", "Oregon", "Texas"])
print(sdata1)
# NaN : 누락값
print(pd.isna(sdata1))
print(pd.notna(sdata1))
print(sdata1.isna())
```

```python
#  연산
print(sdata)
print(sdata1)
print(sdata + sdata1) # index끼리의 연산이 된다.
sdata.name="이숭무" # 시이즈에 이름 주기
sdata.index.name='state' # index에 이름 주기
print(sdata)
```

```python
# index 순서 바꾸기
sdata.index=['Oregon','Utah','Ohio','Texas']
print(sdata)
```

```python
# Reindexing
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])
obj
```

```python
print(obj)
obj2 = obj.reindex(["a", "b", "c", "d", "e"]) # index의 순서 변경
print(obj2)
```

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
print(obj3)
reindexed = obj3.reindex(np.arange(6), method='ffill')
print(reindexed)
reindexed = obj3.reindex(np.arange(6), method='bfill')
print(reindexed)
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=["a", "c", "d"],
                     columns=["Ohio", "Texas", "California"])
print(frame)
frame2 = frame.reindex(index=["a", "b", "c", "d"]) # index를 재정의 할 경우 없던 index에는 값이 NaN이 된다.
print(frame2)
```

```python
print(frame)
print(frame['Ohio'])
print(type(frame['Ohio']))
print(frame.loc['a'])
print(type(frame.loc['a']))
print(frame.loc['a']['Ohio'])
print(frame.loc[['a','c','d']])
print(type(frame.loc[['a','c','d']]))
print(frame.loc[['a','c','d']]['Ohio'])
print(frame.loc[['a','c','d']][['Ohio','Texas']])
```
