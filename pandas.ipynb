```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
data
```

```python
new_data = data.drop(index=["Colorado", "Ohio"]) # DataFrame에서는 행당 index행을 삭제할 수 있다.
print(new_data)
new_data = data.drop(columns=["two"]) # columns 키워드를 이용하여 컬럼을 삭제
print(new_data)
new_data = data.drop("two", axis=1)  # axis=1은 열을 의미 한다. (행기준, y축기준)
print(new_data)
new_data = data.drop(["two", "four"], axis="columns")
print(new_data)
new_data = data.drop(["Ohio", "Colorado"], axis=0) #  axis=0은 행을 의미 (열 기준 , x축 기준)
print(new_data)
new_data = data.drop(["Ohio", "Colorado"], axis="index") #  axis=0은 행을 의미 (열 기준 , x축 기준)
print(new_data)
```

```python
# Indexing, Selection, and Filtering
obj = pd.Series(np.arange(4.), index=["a", "b", "c", "d"])
print(obj)
print(obj["b"]) #명시적 index
print(obj[1]) # 암시적 index
print(obj[2:4]) # 암시적 index를 이용한 slicing
print(obj[["b", "a", "d"]]) # 명시적
print(obj[[1, 3]]) # 암시적
print(obj[obj < 2]) # 부울 index
```

```python
print(obj)
print(obj.loc[["b", "a", "d"]]) # Series에서 loc사용
print(obj.iloc[[1,0,2]])
print(obj.loc["b":"c"]) # 명시적index :  b부터 c까지
print(obj.iloc[1:3]) # 암시적 index :  1부터 3전까지
obj.iloc[1:3] = 5
print(obj)
```

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
print(data)
print(data["two"]) # 열이름으로 데이터 가져오기
print(data[["three", "one"]])
print(data[:2]) # 암시적 index를 사용해서 행 값을 출럭
print(data.iloc[:2]) # 암시적 index를 사용해서 행 값을 출럭
print(data[data["three"] > 5])
data[data < 5] = 0
print(data)
print(data.loc["Colorado"] )
print(data.loc[["Colorado", "New York"]])
print(data.loc['Colorado',["two", "three"]])
# loc, iloc로 데이터를 가져 올 때는  행열로 가져운다.
```

```python
print(data)
print(data.iloc[2]) # 암시적 사용
print(data.iloc[[2, 1]])
print(data.iloc[2, [3, 0, 1]])
print(data.iloc[[1, 2], [3, 0, 1]])
print(data.iloc[:, :3][data.three > 5])
data.loc[data.three >= 2]
```

```python
# 산술 연산과 데이터 정렬
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1],
               index=["a", "c", "e", "f", "g"])
s1, s2
```

```python
s1 + s2 # index가 없으면 index를 맞추기 위해 브로드케스트를 하는데 브로드케스트된 index는 NaN값을 갖느다
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list("bcd"),
                   index=["Ohio", "Texas", "Colorado"])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list("bde"),
                   index=["Utah", "Ohio", "Texas", "Oregon"])
df1, df2
```

```python
df1 + df2 # 행과 열이 같은 것 끼리의 연산이므로 크기가 다르면 브로드케스트를 해서 행과 열을 맞춰준다
```
