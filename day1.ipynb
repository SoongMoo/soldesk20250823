```python
"이숭무의 나이는 20살이고 키는 175.3입니다"
name = "이숭무"
age = 20
height=175.3
print("%s의 나이는 %d살이고 키는 %f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %5.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %3.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %10.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %-10.1f입니다" % (name, age, height))
print("%s의 나이는 %s살이고 키는 %s입니다" % (name, age, height))
```

```python
# format함수를 이용한 포맷팅 : format()
# "a" :문자 , a :변수 , a() : 함수 => print(), int(), str(),
"이숭무의 나이는 20살이고 키는 175.3입니다"
name = "이숭무"
age = 20
height = 175.3
print("%s의 나이는 %d살이고 키는 %f입니다" % (name, age, height))
print("{0}의 나이는 {1}살이고 키는 {2}입니다".format(name, age, height))
#                                                      0     1     2
print("{0}의 나이는 {2}살이고 키는 {1}입니다".format(name, height, age))
print("{0}의 나이는 {2}살이고 키는 {1}입니다".format(name, 175.3, age))
```

```python
# format함수에 있는 이름 사용하기
print("{0}의 나이는 {a}살이고 키는 {1}입니다".format(name, 175.3, a=age))
#                                                      0     1
print("{0}의 나이는 {a}살이고 키는 {h}입니다".format(name, h=175.3, a=age))

#print("{0}의 나이는 {a}살이고 키는 {h}입니다".format(name, h=175.3, age))
# 오류                                                     0
```

```python
print("{0:10}의 나이는 {a:10}살이고 키는 {1:10}입니다".format(name, 175.3, a=age))
#정렬
print("{0:10}의 나이는 {a:10}살이고 키는 {1:10}입니다".format(name, 175.3, a=age))
# 기본값 : 문자(왼쪽) , 숫자(오른쪽)
print("{0:>10}의 나이는 {a:>10}살이고 키는 {1:>10}입니다".format(name, 175.3, a=age)) # 오른쪽 정렬
print("{0:<10}의 나이는 {a:<10}살이고 키는 {1:<10}입니다".format(name, 175.3, a=age)) # 왼쪽 정렬
print("{0:^10}의 나이는 {a:^10}살이고 키는 {1:^10}입니다".format(name, 175.3, a=age)) # 가운데 정렬
print("{0:#^10}의 나이는 {a:-^10}살이고 키는 {1:$^10}입니다".format(name, 175.3, a=age))
```
